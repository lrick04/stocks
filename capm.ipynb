```python
R_f = float(input("Add mega kockázatmentes hozamot! (%)")) / 100
R_m = float(input("Add meg a piaci hozamot! (%)")) / 100
β = float(input("Add meg az eszköz bétáját!"))
CAPM = float((R_f + β*(R_m-R_f)) * 100)

CAPM_kerekítve = round(CAPM, ndigits=4)

print(f"Az eszköz elvárt hozama a CAPM modell alapján {CAPM_kerekítve}%.")

```

```python
import numpy as np
import matplotlib.pyplot as plt

beta_values = np.linspace(0,5,10)

CAPM_SML = (R_f + beta_values*(R_m-R_f)) * 100

plt.figure(figsize=(8,6))
plt.plot(beta_values, CAPM_SML, label = 'Security market line', color = "blue")
plt.title("Security market line (SML)")
plt.xlabel("Beta (β)")
plt.ylabel("Expected return")
plt.grid()
plt.scatter(β, CAPM, color = 'green')
plt.scatter(1, R_m*100, color = 'red')
plt.scatter(1, R_m * 100, color='green', s=80, label=f"Market (β=1, R_m={R_m*100:.2f}%)")
plt.scatter(β, CAPM, color='red', s=80, label=f"Your asset (β={β}, {round(CAPM, 2)}%)")
plt.legend()
plt.show()






```
